```python
import pandas as pd
import numpy as np


df = pd.read_csv("Titanic.csv")
df
```

```python
df.info
```

```python
#Imputation stratergy 

print('Missing value present in age coloumn :', df['age'].isnull().sum() / len (df)*100)
```

```python
df['age'].median(),df['age'].mean()
```

```python
df['age'].fillna(df['age'].median(),inplace = True)
```

```python
print('Missing value present in age coloumn :', df['age'].isnull().sum() / len (df)*100)
```

```python
df['embarked'].fillna('S',inplace = True)
```

```python
print('Missing value present in Embarked coloumn :', df['embarked'].isnull().sum() / len (df)*100)
```

```python
df['embarked'].value_counts()
```

```python
df.isnull().sum().sort_values(ascending = True).plot(kind = 'barh', color = 'pink')./
```

```python
#segragate num and data
categorical_col = df.select_dtypes(include = 'object')
numerical_col = df.select_dtypes(exclude = 'object')
```

```python
# Using Encoding Technique 
#it will convert cate_data into numebers
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()

for i in categorical_col.columns:
    df[i] = le.fit_transform(df[i])
    
df
```

```python
df['alone'] = le.fit_transform(df['alone'])
df
```

```python
df['alone'] = df['alone'].astype(int)
df['alone'].value_counts()
```
