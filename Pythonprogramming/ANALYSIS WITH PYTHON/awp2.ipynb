```python
import pandas as pd
```

```python
db = pd.DataFrame({
    'Country': ['USA', 'India', 'China', 'Russia'],
    'GDP': [28500, 3900, 17600, 1900],
    'Loan': [34000, 2500, 14000, 300],
    'Education investment': [1100, 120, 400, 30],
    'Defence sector investment': [877, 82, 293, 89],
    'Tech Investment': [860, 60, 620, 25],
    'Sales': [2700, 750, 3600, 490],
    'InflationRate': [8, 9, 10, 11]
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(data = db, x= db['Country'], y = db['Loan'] , linestyle = '--', marker = 'o')

plt.show()
```

```python
sns.barplot(data = db, x= db['Country'], y = db['Loan'])

plt.show()
```

```python
sns.countplot(data = db, x= db['Country'])

plt.show()
```

```python
sns.displot(x = db['Tech Investment'], kde = True)
plt.show()
```

```python
sns.scatterplot(data = db , x = db['Education investment'] , y = db['Defence sector investment'])
plt.show()
```

```python
sns.scatterplot(data = db , x = db['Defence sector investment'] , y = db['Sales'], color = 'black' , hue= db['Country'])
plt.show()
```

```python
sns.regplot(data = db , x = db['GDP'] , y = db['Sales'], ci = 10.0, marker = 'o', label = 'Regression line')
plt.ylim(500,5000)
plt.xlim(1000,30000)
plt.legend()
plt.grid()
plt.show()
```

```python
sns.pairplot(data=db)
plt.show()
```

```python
sns.jointplot(data=db, 
              x = db['GDP'], 
              y = db['Sales'], 
              hue=db['InflationRate'])
plt.axhline(7000,linestyle = '--' , color = 'lightblue')
plt.axvline(7,linestyle = '--' , color = 'lightblue')
plt.show()
```
